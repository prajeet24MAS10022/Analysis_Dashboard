```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Generate dates
dates = pd.date_range(datetime.now() - timedelta(days=30), periods=30)

# Create mock data
data = pd.DataFrame({
    'Date': dates,
    'Revenue': np.random.randint(1000, 5000, size=30),
    'Users': np.random.randint(100, 500, size=30),
    'Conversions': np.random.randint(10, 100, size=30)
})
data['Growth %'] = data['Revenue'].pct_change().fillna(0) * 100
data.head()
```

```python
import plotly.express as px

fig = px.line(data, x='Date', y='Revenue', title='Daily Revenue Trend')
fig.show()
```

```python
fig = px.bar(data, x='Date', y='Conversions', title='Conversions per Day')
fig.show()
```

```python
labels = ['Organic', 'Paid', 'Referral', 'Social']
values = [25, 35, 20, 20]

fig = px.pie(names=labels, values=values, title='Conversion Share by Channel')
fig.show()
```
